```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
#swarmplot
sns.swarmplot(x = 'day', y = 'total_bill', data = df, hue = 'sex')
```

```python
sns.catplot(data = df, x = 'day', y = 'total_bill', hue = 'sex', kind = 'strip')
```

```python
df.head()
```

# Regression Plot

```python
sns.regplot(data = df, x = 'total_bill', y = 'tip', color = 'green')
```

# Linear Model Plot

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = df)
```
