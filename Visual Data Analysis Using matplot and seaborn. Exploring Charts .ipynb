```python
Students = ['Femi', 'Funke', 'Temi', 'Tife', 'Amir']
Scores = [20, 50, 30, 70, 50]
```

```python
data = sns.load_dataset('flights')
```

```python
data.head()
```

```python
# How did the total number of passengers change over the months
passStats_trend = data.groupby('year')['passengers'].sum()
```

```python
plt.figure(figsize = (10, 6)) # This specifies the parameters for the Chart, 10 is the lenght and 6 the height
plt.plot(passStats_trend.index, passStats_trend.values, marker = '*')
plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.xticks(passStats_trend.index)
plt.yticks(passStats_trend.values)
plt.title('Total Passengers Across The Years')
plt.show()
```

```python
data.head()
```

```python
# Which month has the highest average number of passengers
MonthAverg = data.groupby('month')['passengers'].mean()
```

```python
MonthAverg
```

```python
plt.bar(MonthAverg.index, MonthAverg.values)
plt.title('Average Number Of Passengers Per Month')
plt.show()
```

```python
storeA = [20, 19, 20, 45, 34, 50, 70, 60, 100, 20, 10, 50]
storeB = [70, 89, 90, 50, 20, 90, 90, 30, 50, 80, 40, 90]
plt.plot(MonthAverg.index, storeA, label = 'Store A', marker = '*')
plt.plot(MonthAverg.index, storeB, label = 'Store B', marker = 'o')
plt.legend # this enables the labeling of the two lines to show
plt.xlabel('Months')
plt.ylabel('Income')
plt.title('Comparing The Income of Two Competing Stores')
plt.show()
```
